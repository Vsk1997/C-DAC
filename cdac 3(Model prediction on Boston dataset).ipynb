```python
def calculate_weight(x,y,epouch,print_interval):
    print("Iteration","Error","Bias","\tWeight",sep="\t\t\t")
    x=np.insert(x,0,1,axis=1)
    theta=np.random.randn(len(x[0]))
    m=y.size
    old_error,error=0,0
    for i in range(epouch):
        h=(x@theta).reshape(-1,1) ##Hadamord product @
        error=((y-h)**2).mean()
        theta=theta-0.05*(1/m)*(x.T.dot(h-y)).reshape(-1)
        print(str(i).center(10),error,theta[0],theta[1],sep="\t\t") if not i%print_interval else _
        if abs(old_error-error)<0.0000000001:
            break
        old_error=error
    print(str(i+1).center(10),error,theta[0],theta[1],sep="\t\t")
    return theta
```

```python
theta=calculate_weight(x,y,50000,2000)
```

```python
def Predict(p,theta):
    x=np.insert(p,0,1,axis=0)
    return x.dot(theta)
```

```python
y_cap=np.array([Predict(i,theta) for i in x])
fig=plt.figure(figsize=(10,7),dpi=100)
ax=fig.add_subplot(111,projection='3d')
ax.set_facecolor("white")
ax.scatter(x[:,2],x[:,7],y,c="r")
ax.scatter(x[:,2],x[:,7],y_cap,c="aqua")
plt.subplots_adjust(top=1.1,bottom=-0.1,right=1,left=-0.2,hspace=0,wspace=0)
```

```python
def goodness_of_fit(x,y,epouch):
    x=np.insert(x,0,1,axis=1)
    theta=np.random.randn(len(x[0]))
    m=y.size
    old_error,error=0,0
    for i in range(epouch):
        h=(x@theta).reshape(-1,1) ##Hadamord product @
        error=((y-h)**2).mean()
        theta=theta-0.05*(1/m)*(x.T.dot(h-y)).reshape(-1)
        if abs(old_error-error)<0.0000000001:
            break
        old_error=error
    sst=np.sum((y-y.mean())**2,axis=0)
    ssr=np.sum((h-y.mean())**2,axis=0)
    sse=np.sum((y-h)**2,axis=0)
    R2=ssr/sst
    return R2
```
